```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.greatplacetowork.com/best-workplaces/100-best/2023"
```

```python
r = requests.get(url)
```

```python
html = str(r.content)
```

```python
print(html)
```

```python
companies = []
industries = []
locations = []
rank = []
```

```python
company_herf = "class=\"link h5\" title="
industry_herf = "<li><i class=\"fa-li fa fa-building-o\" aria-hidden=\"true\"></i>"
location_herf = "<li><i class=\"fa-li fa fa-map-marker\" aria-hidden=\"true\"></i>"
```

```python
i = 0
while(html.find(company_herf) != -1):

    i += 1
    rank.append(str(i))

    index = html.find(company_herf) + len(company_herf) + 1
    html = html[index: ]
    index2 = html.find("\"")
    company = html[0: index2]
    companies.append(company)

    index = html.find(industry_herf) + len(industry_herf)
    html = html[index: ]
    index2 = html.find("<")
    industry = html[0: index2]
    industries.append(industry)

    index = html.find(location_herf) + len(location_herf)
    html = html[index: ]
    index2 = html.find("<")
    location = html[0: index2]
    locations.append(location)


print(len(locations))
```

```python
df = pd.DataFrame({"": rank, "Company": companies, "Industry": industries, "Location": locations})
print(df)
```

```python
df.to_excel("2023_GPTW_top100_results.xlsx", index = False)
```

```python
filename = "GPTW_top100_results.xlsx"

company_herf = "class=\"link h5\" title="
industry_herf = "<li><i class=\"fa-li fa fa-building-o\" aria-hidden=\"true\"></i>"
location_herf = "<li><i class=\"fa-li fa fa-map-marker\" aria-hidden=\"true\"></i>"

with pd.ExcelWriter(filename) as outputFile:
    
    for year in range(2016, 2024):
        url = "https://www.greatplacetowork.com/best-workplaces/100-best/" + str(year)
        r = requests.get(url)
        html = str(r.content)
        
        companies = []
        industries = []
        locations = []
        rank = []

        i = 0
        while(html.find(company_herf) != -1):

            i += 1
            rank.append(str(i))

            index = html.find(company_herf) + len(company_herf) + 1
            html = html[index: ]
            index2 = html.find("\"")
            company = html[0: index2]
            companies.append(company)

            index = html.find(industry_herf) + len(industry_herf)
            html = html[index: ]
            index2 = html.find("<")
            industry = html[0: index2]
            industries.append(industry)

            index = html.find(location_herf) + len(location_herf)
            html = html[index: ]
            index2 = html.find("<")
            location = html[0: index2]
            locations.append(location)

            df = pd.DataFrame({"": rank, "Company": companies, "Industry": industries, "Location": locations})
            df.to_excel(outputFile, sheet_name = str(year), index = False)
        
        
```
